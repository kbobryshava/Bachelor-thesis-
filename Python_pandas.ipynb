# Pandas

```python
import pandas as pd
```

```python
client_pool = pd.read_csv('client_pool.csv')
city_dict = pd.read_csv('city_dict.csv')
```

```python
print(client_pool.describe())
```

```python
client_pool_city=client_pool.groupby('id_city')['id_client'].count().reset_index()
client_pool_gender=client_pool.groupby('gender')['id_client'].count().reset_index()
client_pool_city_gender=client_pool.groupby(['gender','id_city'])['id_client'].count().reset_index()
```

```python
client_pool_city_f=pd.merge(client_pool,city_dict)
client_pool_city_f.drop(columns='id_city', inplace=True)
```

```python
client_pool_city_gender_f=pd.merge(client_pool_city_gender,city_dict)
client_pool_city_gender_f.drop(columns='id_city', inplace=True)
client_pool_city_gender_f
```

```python
client_pool_city_gender_f_women = client_pool_city_gender_f.loc[client_pool_city_gender_f['gender'] == 'F'].copy()
client_pool_city_gender_f_women.rename(columns={'id_client': 'cnt_women'}, inplace=True)
client_pool_city_gender_f_women
```

```python
client_pool_city_gender_f_men=client_pool_city_gender_f.loc[client_pool_city_gender_f['gender'] == 'M'].copy()
client_pool_city_gender_f_men.rename(columns={'id_client': 'cnt_men'}, inplace=True)
client_pool_city_gender_f_men
```

```python
client_pool_city_gender_gencol=pd.merge(client_pool_city_gender_f_women, client_pool_city_gender_f_men,  on='name_city', how='outer')
client_pool_city_gender_gencol['cnt_all']=client_pool_city_gender_gencol['cnt_women']+client_pool_city_gender_gencol['cnt_men']
client_pool_city_gender_gencol = client_pool_city_gender_gencol[['name_city', 'cnt_men', 'cnt_women', 'cnt_all']]
client_pool_city_gender_gencol 
```

```python
client_pool_city_gender_gencol['share_men']=client_pool_city_gender_gencol['cnt_men']/client_pool_city_gender_gencol['cnt_all']
client_pool_city_gender_gencol['share_women']=client_pool_city_gender_gencol['cnt_women']/client_pool_city_gender_gencol['cnt_all']
row_with_highest_share_of_men = client_pool_city_gender_gencol[client_pool_city_gender_gencol['share_men'] == client_pool_city_gender_gencol['share_men'].max()]
row_with_highest_share_of_women = client_pool_city_gender_gencol[client_pool_city_gender_gencol['share_women'] == client_pool_city_gender_gencol['share_women'].max()]
print(row_with_highest_share_of_men)
print(row_with_highest_share_of_women)
```
