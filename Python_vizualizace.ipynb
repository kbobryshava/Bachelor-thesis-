# Příklad vizualizace v Python 

```python
from matplotlib import pyplot as plt
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
plt.scatter(x, y, color='red', edgecolor = 'blue')
plt.title('Bodový graf')
plt.xlabel('Osa Х')
plt.ylabel('Osa Y')
plt.grid()
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
plt.figure(figsize=(7, 7))
plt.plot(x, y, 'g')
plt.xlabel('osa Х')
plt.ylabel('osa Y')
plt.title('Čárový graf')
plt.vlines(6, ymin=0, ymax=30, color='red', label = 'Maximum')
plt.vlines(2, ymin=0, ymax=30, color='black', linestyle='--', label = 'Minimum')
plt.grid()
plt.legend(loc = 'lower right')
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
z=[6,9,13,13,16,12,20,18,17]
plt.figure(figsize=(7, 7))
plt.plot(x, y, 'g')
plt.plot(x, z, 'm-', markersize=1)
plt.xlabel('osa Х')
plt.ylabel('osa Y')
plt.title('Graf s dvěma čárami')
plt.vlines(6 - 0.03, ymin=0, ymax=30, color='red', linestyle='--', label = 'Maximum Х')
plt.vlines(2, ymin=0, ymax=30, color='deepskyblue', label = 'Minimum Х')
plt.vlines(6 + 0.03, ymin=0, ymax=30, color='blue', linestyle='--', label = 'Maximum Z')
plt.vlines(1, ymin=0, ymax=30, color='salmon', label = 'Minimum Z')
plt.legend(loc = 'lower right')
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
z=[6,9,13,13,16,12,20,18,17]
fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(7, 7))
ax1.plot(x, y, 'g')
ax1.vlines(6 - 0.03, ymin=0, ymax=30, color='red', linestyle='--', label = 'Maximum Х')
ax1.vlines(2, ymin=0, ymax=30, color='deepskyblue', label = 'Minimum Х')

ax11 = ax1.twinx()
ax11.plot(x, z, 'm-', markersize=1)
ax11.vlines(6 + 0.03, ymin=0, ymax=30, color='blue', linestyle='--', label = 'Maximum Z')
ax11.vlines(1, ymin=0, ymax=30, color='salmon', label = 'Minimum Z')

plt.xlabel('osa Х')
plt.ylabel('osa Y')
plt.legend(loc = 'lower right')
plt.title('Stejný graf s využitím sublopts')
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
z=[6,9,13,13,16,12,20,18,17]
difference = [y_val - z_val for y_val, z_val in zip(y, z)]

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax1.plot(x, y, 'g')
ax1.vlines(6 - 0.03, ymin=0, ymax=30, color='red', linestyle='--', label = 'Max Х')
ax1.vlines(2, ymin=0, ymax=30, color='deepskyblue', label = 'Min Х')
ax1.set_ylabel('osa Y')

ax11 = ax1.twinx()
ax11.plot(x, z, 'm-', markersize=1)
ax11.vlines(6 + 0.03, ymin=0, ymax=30, color='blue', linestyle='--', label = 'Max Z')
ax11.vlines(1, ymin=0, ymax=30, color='salmon', label = 'Min Z')
ax11.yaxis.set_tick_params(direction='out', labelright=False)

lines1, labels1 = ax1.get_legend_handles_labels()
lines11, labels11 = ax11.get_legend_handles_labels()
ax1.legend(lines1 + lines11, labels1 + labels11,loc = 'lower right')

ax12 = ax1.twinx()
ax12.bar(x, difference, alpha = .3, color = 'green', label = 'Rozdíl')
ax12.yaxis.set_tick_params(labelright=True, direction='out')
ax12.legend(loc = 'upper right')
plt.grid()
plt.title('Rozdíl mezi hodnotami X a Z')
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8,9]
y=[8,5,6,10,18,26,18,22,23]
z=[6,9,13,13,16,12,20,18,17]
difference = [y_val - z_val for y_val, z_val in zip(y, z)]

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(10, 6))
ax1.plot(x, y, 'g')
ax1.vlines(6 - 0.03, ymin=0, ymax=30, color='red', linestyle='--', label = 'Max Х')
ax1.vlines(2, ymin=0, ymax=30, color='deepskyblue', label = 'Min Х')
ax1.set_ylabel('Osa Y')

ax11 = ax1.twinx()
ax11.plot(x, z, 'm-', markersize=1)
ax11.vlines(6 + 0.03, ymin=0, ymax=30, color='blue', linestyle='--', label = 'Max Z')
ax11.vlines(1, ymin=0, ymax=30, color='salmon', label = 'Min Z')
ax11.yaxis.set_tick_params(direction='out', labelright=False)


ax12 = ax1.twinx()
ax12.bar(x, difference, alpha = .3, color = 'green', label = 'Rozdíl')
ax12.yaxis.set_tick_params(labelright=True, direction='out')
ax12.set_ylim(0, 15)
lines1, labels1 = ax1.get_legend_handles_labels()
lines11, labels11 = ax11.get_legend_handles_labels()
lines12, labels12 = ax12.get_legend_handles_labels()
ax1.legend(lines1 + lines11 + lines12, labels1 + labels11 + labels12,loc = 'best')

plt.grid()
plt.title('Absolutní rozdíl mezi hodnotami X a Z')
plt.show()

```

```python
import pandas as pd
import datetime as dt
months = ['Červen','Červenec','Srpen','Září','Řijen','Listopad','Prosinec']
purchase_list=pd.read_csv('Purchase_List.csv')
purchase_list['date_purchase'] = pd.to_datetime(purchase_list['date_purchase'], format = '%m/%d/%Y')
purchase_list['month_purchase']=purchase_list['date_purchase'].dt.month
purchase_list_gr=purchase_list.groupby('month_purchase').agg({'id_purchase':'count', 'amt_pay':'sum'}).reset_index()
purchase_list_gr.rename(columns={'id_purchase': 'cnt_purchase'}, inplace=True)
purchase_list_gr.rename(columns={'amt_pay': 'amt_sum'}, inplace=True)

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(6, 6))
ax1.plot(purchase_list_gr['month_purchase'], purchase_list_gr['amt_sum'], 'b', label = 'Celkový šek')
ax11 = ax1.twinx()
ax11.bar(purchase_list_gr['month_purchase'], purchase_list_gr['cnt_purchase'], alpha = .1, color = 'green', label = 'Počet plateb')

ax1.set_title('Platby celkem')
ax1.legend(loc = 'lower left')
ax11.legend(loc = 'upper right')
ax1.set(xticks=purchase_list_gr['month_purchase'], xticklabels=months)
ax1.xaxis.set_tick_params(rotation=45)
plt.grid()
plt.show()
```

```python
purchase_list_gr['avg_bill']=purchase_list_gr['amt_sum']/purchase_list_gr['cnt_purchase']
purchase_list_gr['avg_bill'] = purchase_list_gr['avg_bill'].round(1)
purchase_list_gr.reset_index()

x = purchase_list_gr['month_purchase']
y = purchase_list_gr['avg_bill']
plt.scatter(x, y, color='red', edgecolor = 'blue')
plt.xticks(purchase_list_gr['month_purchase'], months, rotation=45)
plt.title('Průměrná platba')
plt.ylim(0, 6400)
plt.grid()
plt.show()
```
